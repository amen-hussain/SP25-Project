```python
import pandas as pd
import numpy as np
#import statsmodels.api as sm
import matplotlib.pyplot as plt
import csv 
from sklearn.model_selection import train_test_split  # For splitting the dataset into training and testing sets
from sklearn.linear_model import LinearRegression  # Linear regression model
from sklearn.metrics import mean_squared_error, r2_score  # Evaluation metrics
import matplotlib.pyplot as plt  # For data visualization
```

```python
diabetes = pd.read_csv('/workspaces/SP25-Project/data/diabetic_data.csv')
```

```python
diabetes.replace('?', np.nan, inplace=True)
```

```python
diabetes.drop_duplicates(inplace=True)
```

```python
# Define the columns to drop
columns_to_drop = [
    'encounter_id', 'patient_nbr', 'weight',
    'admission_type_id', 'discharge_disposition_id', 'admission_source_id',
    'payer_code', 'medical_specialty', 'diag_1', 'diag_2', 'diag_3',
    'metformin', 'repaglinide', 'nateglinide', 'chlorpropamide',
    'glimepiride', 'acetohexamide', 'glipizide', 'glyburide', 'tolbutamide',
    'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone',
    'tolazamide', 'examide', 'citoglipton', 'insulin',
    'glyburide-metformin', 'glipizide-metformin',
    'glimepiride-pioglitazone', 'metformin-rosiglitazone',
    'metformin-pioglitazone', 'change', 'diabetesMed'
]

# Drop the columns from the 'diabetes' DataFrame
diabetes_cleaned = diabetes.drop(columns=columns_to_drop)

# Show the first few rows of the cleaned DataFrame
diabetes_cleaned.head(10)
```

```python
# Define the mapping from age ranges to numeric midpoints
# Step 3: Convert age ranges to numerical midpoints
age_mapping = {
    '[0-10)': 5,
    '[10-20)': 15,
    '[20-30)': 25,
    '[30-40)': 35,
    '[40-50)': 45,
    '[50-60)': 55,
    '[60-70)': 65,
    '[70-80)': 75,
    '[80-90)': 85,
    '[90-100)': 95
}
diabetes_cleaned['age'] = diabetes_cleaned['age'].map(age_mapping)
# Preview to confirm
diabetes_cleaned[['age']].head(20)
```

```python
# Step 4: Convert 'readmitted' column to binary
diabetes_cleaned['readmitted'] = diabetes_cleaned['readmitted'].apply(lambda x: 1 if x == '<30' else 0)

# Step 5: Encode 'gender' as binary (Male: 1, Female: 0)
diabetes_cleaned['gender'] = diabetes_cleaned['gender'].replace({'Male': 1, 'Female': 0})

# Step 6: One-hot encode other categorical features
diabetes_cleaned = pd.get_dummies(
    diabetes_cleaned,
    columns=['race', 'max_glu_serum', 'A1Cresult'],
    drop_first=True
)

# Step 7: Fill missing numeric values with the column mean
numeric_cols = diabetes_cleaned.select_dtypes(include=['float64', 'int64']).columns
diabetes_cleaned[numeric_cols] = diabetes_cleaned[numeric_cols].fillna(diabetes_cleaned[numeric_cols].mean())

# Step 8: Data preview
diabetes_cleaned.head(20)
```

```python
print(diabetes_cleaned.columns)
```

```python
# Step 9: Convert 'True' and 'False' values to Bianary 
diabetes_cleaned = diabetes_cleaned.replace({True: 1, False: 0})
```
